```python
import re

with open('/Users/schmorls/Desktop/psEval/deepeval_results/plainData-6500.txt', 'r') as file:
    content = file.read()

scores_strings = re.findall(r'score: (\d+\.\d+)', content)

scores = [float(score) for score in scores_strings]

print("Found: " + str(len(scores)) + " scores")
# Print the filtered scores
print(sum(scores) / 40)
```
