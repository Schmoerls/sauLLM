```python
model_name = "plainData-finetuning-5000"
results_directory = "generation_results/"
correctness_questions_path = "correctness_questions.json"
correctness_answers_path = "correctness_answers.json"

with_logging = True

trained_results_path = results_directory + "correctness_" + model_name + ".json"
print(trained_results_path)
```

```python
import os
import json
from deepeval.test_case import LLMTestCase
from deepeval.dataset import EvaluationDataset
from deepeval.metrics import GEval
from deepeval.test_case import LLMTestCaseParams

os.environ["OPENAI_API_KEY"] = "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"

if with_logging: print("Reading correctness questions from: " + correctness_questions_path)
with open(correctness_questions_path, 'r') as file:
    json_data = json.load(file)

correctness_questions = [item for item in json_data]
if with_logging: print(len(correctness_questions))


if with_logging: print("Reading correctness answers from: " + correctness_answers_path)
with open(correctness_answers_path, 'r') as file:
    json_data = json.load(file)

correctness_answers = [item for item in json_data]
if with_logging: print(len(correctness_answers))


trained_results_path = results_directory + "correctness_" + model_name + ".json"
if with_logging: print("Reading trained model Correctness results from: " + trained_results_path)
with open(trained_results_path, 'r') as file:
    json_data = json.load(file)

trained_answers = [item for item in json_data]
if with_logging: print(len(trained_answers))

test_cases = [
    LLMTestCase(input=correctness_question,
                actual_output=trained_answer,
                expected_output=expected_answer)
    for correctness_question, trained_answer, expected_answer in zip(correctness_questions, trained_answers, correctness_answers)
]

print(len(test_cases))

dataset = EvaluationDataset(test_cases=test_cases)

correctness_metric = GEval(
    name="Correctness",
    model="gpt-4o",
    evaluation_params=[
        LLMTestCaseParams.INPUT,
        LLMTestCaseParams.ACTUAL_OUTPUT,
        LLMTestCaseParams.EXPECTED_OUTPUT],
    evaluation_steps=[
        "'expected_output' ist als Musterlösung zu sehen. Beurteile, ob 'actual_output die Frage 'Input' korrekt beantwortet und vergebe Extrapunkte, wenn die gleichen Inhalte wie in der Lösung vorkommen."
        "Vergleiche die Fakten im 'actual output' mit den Fakten 'expected output' und beurteile, wie gut der 'actual output' den 'expected output' widerspiegelt",
        "Überprüfe, ob die Fakten im 'actual output' im Widerspruch zu den Fakten im 'expected output' stehen",
    ],
)

evaluation_output = dataset.evaluate([correctness_metric])
print(evaluation_output)
```
